```python
import numpy as np

all_files = [f for f in os.listdir(os.getcwd())]#输出根path下的所有文件名到一个列表中
#对各个文件进行处理
interdf=pd.DataFrame(columns=['NAME','ADDRESS','TELEPHONE','PROVINCE','CITY','COUNTY','CODE','LON','LAT','TYPECODE','BASETYPE','SUBTYPE','CATEGORY']) 
csv_files=[]
for i in all_files:
    if i.split('.')[1]=='csv':
        if '2014' in i.split('.')[0]:
            name=i.replace('2014pois.csv',"")
            csv_files.append(i)#os.path.join(os.getcwd(),
            citynames=['NAME','ADDRESS','TELEPHONE','PROVINCE','CITY','COUNTY','CODE','LON','LAT','TYPECODE','BASETYPE','SUBTYPE','CATEGORY']
            new_df=pd.read_csv(i,engine='python',encoding='utf_8_sig',error_bad_lines=False,names=citynames,index_col=False)
            
            #进行筛选，不等于本城市且有值
            if new_df.shape[0]>0:
                try:
                    new_df['area'] = new_df.apply(lambda x: getlocation(x['TELEPHONE']), axis=1)
                    new_df['area'].fillna(value="", inplace=True)
                    wh_df=new_df[new_df['area']!=""]
                    
                    if wh_df.shape[0]>0:
                        nowcity=new_df.loc[0,'CITY']
                        #这部分修改
                        wh_df[wh_df['area']!=nowcity]
                        #wh_df['area'] not in nowcity
                    interdf=pd.concat([interdf,wh_df], ignore_index=True) 
                except Exception as e:
                    print(e)
            else:
                pass
```
